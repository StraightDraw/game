The quickest method for solving $2\times 2$ matrix games with mixed strategy solutions (MSS) is using <span style="color:blue">**oddments**</span>.

$$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrrccc}&A&B&&\text{Row Diffs}&&\text{Oddments}\\\hline A&3&-3&&|3-(-3)|=6&\searrow&\frac{4}{10}\\B&-3&1&&|-3-(1)|=4&\nearrow&\frac{6}{10}\end{array}\end{array}$$
