We will investigate a genre of games called <span style='color:blue'><b>Police versus Guerrillas</b></span>. The idea is that an insurgency (the guerrilla force) is striking against the government (the police). The police have two arsenals filled with weapons and amunition, and the guerrillas would like to take either one of them, or both, if possible.

## Units

The police have $p$ units they can send to defend the arsenals while the guerrillas have $g$ units with which they can attack. Since $p\geq 2g$ implies the **police always win** and $p\leq g$ implies the guerrillas always have a winning strategy, we generally are sovling games where

$$g \leq p < 2g$$

Common scenarios would have us analyzing 5P vs 3G, 7P vs 5G, or 8P vs 7G.
