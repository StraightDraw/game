We will investigate a genre of games called <span style='color:blue'><b>Police versus Guerrillas</b></span>. The idea is that an insurgency (the guerrilla force) is striking against the government (the police). The police have two arsenals filled with weapons and amunition, and the guerrillas would like to take either one of them, or both, if possible.

## Game Play

The police defend $2$ arsenals and the guerrillas win if they can take over either one, or both. The two players know at nightfall when the attack(s) will take place, but the police do not know how many attackers will be which locations. The guerrillas do not know how the police will arrange their forces.

<span style='color:blue'><b>Key:</b></span> The police **hold** an arsenal if an equal number of attacking and defending units engage there.

**Suppose the police have $4$ units**. For a strategy such as 

<center><span style='color:blue'>$\textbf{3 - 1}$</span></center></br>

they will randomize their choice of which arsenal to send their greater force to protect. 

**If the guerrillas have $3$ units** with which they can attack, they may play a strategy like

<center><span style='color:blue'>$\textbf{2 - 1}$</span></center></br>

As do the police, the guerrillas will randomly select to which of the arsenals they will the greater force.

## Units

We must decide before analyzing outcomes the configuration of forces for both players. The police will have $p$ units with which they may defend the arsenals while the guerrillas will have $g$ units with which they can attack. Since $p\geq 2g$ implies the **police always win** by dividing their forces equally, and $p\leq g$ implies the guerrillas always have a winning strategy (send forces to one arsenal), we must only solve games where

$$g \leq p < 2g$$

Common scenarios would have us analyzing 
- 5P vs 3G
- 7P vs 5G, or
- 8P vs 7G 

## Example: 3 Police versus 2 Guerrillas

The police with $3$ units have only two distinct strategies:
- $\textbf{3 - 0}$
- $\textbf{2 - 1}$

The strategies of $\textbf{0 - 3}$ and $\textbf{2 - 1}$ are just permutations of the above strategies that will randomize in exactly the same way. Therefore, we only need to analyze the two shown in the bullets above.

The guerrillas with $2$ units have two distinct stragegies:
- $\textbf{2 - 0}$
- $\textbf{1 - 1}$

Hence, the game they are playing can be shown in matrix form:

$$\begin{array}{cc}&\textbf{Guerrillas}\\\textbf{Police}& \begin{array}{r|cc} & \textbf{2 - 0} & \textbf{1 - 1}\\ \hline \textbf{3 - 0} & \frac{1}{2} & 0 \\ \textbf{2 - 1} & 1 & 1 \end{array} \end{array}$$

Notice the outcome of $\textbf{3 - 0}$ versus $\textbf{2 - 0}$ is an expected value. Half the time, the units will be sent to the same arsenal, and the police will win. When the units are sent to different arsenals, the guerrillas win. Thus,
$$E(X) = \frac{1}{2}(0) + \frac{1}{2}(1) = \frac{1}{2}$$

You can verify the other three outcomes are the same regardless of the random location to which the forces are sent. The analysis is straightforward once we have constructed the matrix game. Police $\textbf{2 - 1}$ dominates $\textbf{3 - 0}$, so the Police will always play it. You should verify the guerrillas also have a dominated strategy. Thus, the equilibrium of the game is
$$\begin{align}\textbf{Police: }&\textbf{ 2 - 1}\\\textbf{Guerrillas: }&\textbf{ 1 - 1}\\v &= 1\end{align}$$

## Example: 5 Police versus 4 Guerrillas

Reasoning as above, the game matrix for these forces is as follow:

$$\begin{array}{cc}&\textbf{Guerrillas}\\\textbf{Police}& \begin{array}{r|ccc} & \textbf{4 - 0} & \textbf{3 - 1} & \textbf{2 - 2}\\ \hline \textbf{5 - 0} & \frac{1}{2} & 0 & 0 \\ \textbf{4 - 1} & \frac{1}{2} & \frac{1}{2} & 0 \\ \textbf{3 - 2} & 0 & 1 & 1\end{array} \end{array}$$

You should work through each strategy pair to ensure you know how to do the calcuations of the expected values correctly.

### Domination

The $\textbf{Police : 5 - 0}$ strategy is dominated as is the $\textbf{Guerrillas : 3 - 1}$ strategy. Thus, the game reduces to the following:

$$\begin{array}{cc}&\textbf{Guerrillas}\\\textbf{Police}& \begin{array}{r|cc} & \textbf{4 - 0} & \textbf{2 - 2}\\ \hline \textbf{4 - 1} & \frac{1}{2} & 0 \\ \textbf{3 - 2} & 0 & 1\end{array} \end{array}$$
