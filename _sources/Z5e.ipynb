# Exercises

:::{sidebar} Discovering Game Theory
This course uses an **online textbook**: [Game Theory, a Discovery Approach](https://nordstrommath.com/IntroGameTheory2e/intro_game_theory.html), by Jennifer Nordstrom.
:::

Below, we have several practice problems and solutions. All of them can be solved by starting with <span style="color: blue;"><b>Reduction by Dominance</b></span>. They will reduce to, typically, a $3\times 3$ matrix game that can be solved by <span style="color: purple;"><b>Equalized Expectations</b></span>.

:::{admonition} Exercise 1
Find the **mixed strategy solution** by reducing the matrix game by dominance and then solving by equalized expectations.

$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrr}&A&B&C\\ \hline A&-1&9&4\\ B&6&-4&-4\\ C&3&0&2\\ D&-3&-1&-4\\\end{array}\end{array}$
:::

:::{dropdown} Solution

$$\begin{align*} \vec r = \left[\begin{array}{c}\frac{4}{15} \\\frac{1}{15} \\\frac{2}{3} \\0\end{array}\right] &, \hspace{5mm} \vec c = \left[\begin{array}{c}\frac{46}{75} \\\frac{6}{25}\\\frac{11}{75}\end{array}\right] \\[3mm]v&=\frac{32}{15} \end{align*}$$
:::

:::{admonition} Exercise 2
Find the **mixed strategy solution** by reducing the matrix game by dominance and then solving by equalized expectations.

$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrrr}&A&B&C&D\\ \hline A&-4&2&12&-1\\ B&6&12&8&0\\ C&11&12&1&7\\ D&1&7&6&4\\\end{array}\end{array}$
:::

:::{dropdown} Solution

$$\begin{align*} \vec r = \left[\begin{array}{c}\frac{10}{77}\\0\\\frac{3}{7} \\\frac{34}{77}\end{array}\right] &, \hspace{5mm} \vec c = \left[\begin{array}{c}\frac{1}{11} \\0\\\frac{5}{11} \\\frac{5}{11}\end{array}\right] \\[3mm]v&=\frac{51}{11} \end{align*}$$
:::

:::{admonition} Exercise 3
Find the **mixed strategy solution** by reducing the matrix game by dominance and then solving by equalized expectations.

$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrrr}&A&B&C&D\\ \hline A&1&8&12&6\\ B&2&8&5&5\\ C&5&2&10&12\\ D&7&2&-1&-4\\\end{array}\end{array}$
:::

:::{dropdown} Solution

$$\begin{align*} \vec r = \left[\begin{array}{c}0 \\\frac{2}{5}\\\frac{3}{10}\\\frac{3}{10}\end{array}\right] &, \hspace{5mm} \vec c = \left[\begin{array}{c}\frac{48}{85} \\\frac{31}{85} \\0 \\\frac{6}{85}\end{array}\right] \\[3mm]v&=\frac{22}{5} \end{align*}$$
:::

:::{admonition} Exercise 4
Find the **mixed strategy solution** by reducing the matrix game by dominance and then solving by equalized expectations.

$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrrr}&A&B&C&D\\ \hline A&-1&-4&8&10\\ B&11&-2&4&7\\ C&8&8&2&4\\ D&10&7&6&0\\\end{array}\end{array}$
:::

:::{dropdown} Solution

$$\begin{align*} \vec r = \left[\begin{array}{c}\frac{19}{72} \\0 \\\frac{35}{72} \\\frac{1}{4}\end{array}\right] &, \hspace{5mm} \vec c = \left[\begin{array}{c}0 \\\frac{1}{3} \\\frac{3}{8} \\\frac{7}{24}\end{array}\right] \\[3mm]v&=\frac{55}{12} \end{align*}$$
:::
