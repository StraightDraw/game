# Equalized Expectations

:::{sidebar} Discovering Game Theory
This course uses an **online textbook**: [Game Theory, a Discovery Approach](https://nordstrommath.com/IntroGameTheory2e/intro_game_theory.html), by Jennifer Nordstrom.
:::

For an $n\times n$ matrix game with no PSS, we may use the idea of equalized expectations to solve for optimal strategy mixtures algebraically. Recall that, in the $3\times 3$ case, we have

$$\vec r = \left[\begin{array}{r}r_1\\r_2\\r_3\end{array}\right]\hspace{5mm}\text{and}\hspace{5mm}\vec c = \left[\begin{array}{r}c_1\\c_2\\c_3\end{array}\right]$$

where

$$\begin{align}0\leq r_1,r_2,r_3 &\leq 1\\[3mm]r_1 + r_2 + r_3 &= 1\end{align}$$ 

Also, the same is true for $c_1,c_2,$ and $c_3$.

:::{admonition} Example 1
Given the following matrix game, first verify there is no reduction by dominance. Then solve for the optimal strategy mixtures for Rose and for Colin. What is the value of the game?

$$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrr}&A&B&C\\\hline A&1&-1&6&\\B&9&3&-2&\\C&0&6&-2&\end{array}\end{array}$$
:::

We consider the two values $c_1$ and $c_2$ since

$$c_3 = 1 - c_1 - c_2$$

We are thus trying to solve the example game:

$$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrc}&Ac_1&Bc_2&C(1-c_1-c_2)\\\hline A&1&-1&6&\\B&9&3&-2&\\C&0&6&-2&\end{array}\end{array}$$

where Colin ensures that:

- Strategy A is played $c_1$ proportion of the time,
- Strategy B is played $c_2$ proportion of the time, and
- Strategy C is played $1-c_1-c_2$ proportion of the time.

Then, we have the three expected value calculations:

$$\begin{align}E(\text{Rose} A) = c_1 - c_2 + 6(1-c_1-c_2) &= 6 -7c_2\\[2mm]E(\text{Rose} B) = 9c_1 + 3c_2 - 2(1-c_1-c_2)&=11c_1 + 5c_2 -2\\[2mm]E(\text{Rose} C) = (0)c_1 + 6c_2 - 2(1-c_1-c_2)&=2c_1+8c_2-2\end{align}$$

We can create a linear system of two equations and two unknowns by setting the first and second equations equal to one another and the second and third equations equal to each other.

$$\begin{align}6 -7c_2&=11c_1 + 5c_2 -2\\[2mm]11c_1 + 5c_2 -2&=2c_1+8c_2-2\end{align}$$

Moving terms to the same side, we find the following:

$$\begin{align}11c_1 + 12c_2 -8&=0\\[2mm]9c_1 - 3c_2&=0\end{align}$$


$\text{Rose} = (1/2, 1/6, 1/3)$

$\text{Rose} = (2/13, 6/13, 5/13)$
