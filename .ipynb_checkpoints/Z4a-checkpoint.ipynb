# Exercises

:::{sidebar} Discovering Game Theory
This course uses an **online textbook**: [Game Theory, a Discovery Approach](https://nordstrommath.com/IntroGameTheory2e/intro_game_theory.html), by Jennifer Nordstrom.
:::

Below, we have several practice problems and solutions. All of them can be solved by using Reduction by Dominance. Individual problems will indicate whether a complete solution is possible using only reduction by dominance.

:::{admonition} Exercise 1
Find the equilibrium pair by reducing the matrix game by dominance.

$\begin{array}{cc}&\text{Colin}\\\text{Rose}&\begin{array}{r|rrr}&A&B&C\\ \hline A&-2&-4&10\\ B&-1&2&7\\ C&5&5&9\\\end{array}\end{array}$
:::

:::{dropdown} Solution

$$\begin{align*} \text{Equilibrium Pair} &: (\text{Rose }C,\text{Colin }B) \\[3mm]v&=5 \end{align*} $$
:::
