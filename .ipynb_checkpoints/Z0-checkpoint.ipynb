What is the nature of game theory? What are we to accomplish? **What is winning** in a game theory game? Our textbook launches with a quote that will help us see to the heart of the matter. 

:::{important}
Game Theory is not about *playing games*. It is about conflict resolution among **rational** but **distrustful** beings (emphasis added). </br>
                      - - Poundstone
:::

Nordstrom provides an excellent introduction which should be read. She argues that, while games are often enjoyable recreation, even simple games can model real-world situations such as social, economic or even political scenarios.

:::{tip}
Scientists can utilize game theory once its structure is developed. They can study evolutionary biology, philosophy, missile defense, nuclear weapons proliferation, and economics.
:::

The key to games are the distrustful but rational players.

## Rational Play Principle

:::{sidebar} Assignment
Read Chapter 1 in our textbook which starts 
<a href="https://nordstrommath.com/IntroGameTheory2e/intro_game_theory-4.html" target="_blank">here</a> and continues through Sections 1.1 and 1.2. 
:::

The rational play principle is introduced by Nordstrom as having [two components](https://nordstrommath.com/IntroGameTheory2e/S_intro.html):

1. **Players are <span style="color: purple;">self-interested</span>.** The goal is to win the most or lose the least.
2. **Players are <span style="color: purple;">perfectly rational</span>.** A player will always take into account all available information and make the decision which maximizes his payoff. This includes knowing that his opponent is also making the best decision for herself.

The two key player qualities are **rationality** and **self-interest**. Together, these qualities create the conflict we will try to resolve in a way that is satifactory to both players.

## Players

:::{sidebar} Assignment
Be sure to do the **Check Your Understanding** questions for section 1.1 at <a href="https://nordstrommath.com/IntroGameTheory2e/S_intro.html#S_intro-5" target="_blank">this link</a>.
:::

While we generally think of human beings as the players in our games, please note that players are the actors. For example, we can have a <span style="color: purple;">game against nature</span> where nature (the winds or tides or rain) is the actor. The stock market is sometimes modeled as a game againsts nature. We can model **black jack** as a game against nature where the house (or casino) plays by certain rules and is governed largely by probability.

We begin by modeling games that humans play although, at times, chance will enter into the game scheme and affect our decisions. **Poker** is an example of such a game where the cards dealt to each player will influence their choice of strategies. While cooperation is possible amongst players of a game, to begin we will assume that the interests of the human players are <span style="color: purple;"><b>perfectly opposed</b></span>.
